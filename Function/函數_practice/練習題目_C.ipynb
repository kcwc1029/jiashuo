### Problem. 網購免運費判斷
網購平台規定：購物車滿 1000 元（含）即享免運費；若未滿 1000 元，則需加收 60 元的物流運費。請寫一個函數幫結帳系統算出顧客最終需要支付的金額。

* 函數名稱：calculate_checkout
* 輸入變數：amount (購物車總額，整數)

輸出：回傳加上運費（如果有的話）的最終結帳金額 (整數)。

### Problem. 飲料店快速點餐系統
飲料店的 POS 系統需要記錄客人的飲料與冰塊甜度。因為大部分客人都是點「正常冰、正常糖」，請設計一個函數，使用「預設參數」，讓店員在客人沒特別指定時，可以少輸入一些字。

* 函數名稱：order_drink
* 輸入變數：drink_name (飲料名稱，字串)、ice (冰塊，字串，預設為 "正常冰")、sugar (甜度，字串，預設為 "正常糖")

輸出：回傳一段確認字串，格式為："您點的是[飲料名稱]，[冰塊]，[甜度]。"

### Problem. 尋找班上最高分
老師想在期末考後，快速找出班上的最高分是幾分，以便決定是否要調整成績。請設計一個函數來找出串列中的最大值。（提示：可以使用 Python 內建函數或使用 for 迴圈逐一比較）

* 函數名稱：find_highest
* 輸入變數：scores (包含多個成績的串列 list)

輸出：回傳串列中的最高分數 (整數)。

### Problem. 網站密碼強度檢查員
網站註冊系統需要檢查使用者設定的密碼長度，為了安全起見，如果密碼長度少於 8 個字元，系統要拒絕註冊。
* 函數名稱：check_password
* 輸入變數：password (使用者輸入的密碼，字串)

輸出：若密碼長度大於或等於 8，回傳 True；否則回傳 False (布林值)。

### Problem. 餐廳分帳計算機 (多個回傳值)
一群朋友去吃熱炒，結帳時需要把總金額平分。但結帳時只能付整數（小數點無條件捨去），剩下的零頭由當天請客的壽星吸收。請設計一個函數計算每個人要付多少錢，以及壽星要幫忙出的零頭。

* 函數名稱：split_bill
* 輸入變數：total_amount (總金額，整數)、people (人數，整數)
* 輸出：同時回傳兩個值：每個人平均應付的金額 (整數)、剩下的零頭 (整數)。
* 提示：可以使用整數除法 // 與 取餘數 %。

### Problem. 及格人數統計器
助教拿到了一份平時測驗的成績單，他需要統計這次考試中，成績及格（大於或等於 60 分）的同學總共有幾位。

* 函數名稱：count_pass
* 輸入變數：scores (包含多個成績的串列 list)

輸出：回傳及格的「總人數」 (整數)。

### Problem. 停車費計算機
停車場的計費規則如下：不論停多久，基本費用就是 50 元；但如果停車時間超過 2 小時（120 分鐘），超過的部分每分鐘加收 1 元。請寫一個函數幫繳費機計算費用。

* 函數名稱：parking_fee
* 輸入變數：minutes (總停留分鐘數，整數)
輸出：回傳總停車費用 (整數)。
