### Problem. 每日記帳無上限加總
有時候我們一天會買很多零碎的東西（例如：買咖啡、買文具、繳水費）。記帳App需要一個功能，讓使用者可以一次輸入當天「任意數量」的花費金額，並計算出今天的總支出。
* 函數名稱：daily_expenses
* 輸入變數：expenses (任意數量的花費金額，整數)
* 輸出：回傳當日的總花費金額 (整數)。若沒有傳遞任何數值，則回傳0
<!--
```python=
def daily_expenses(*expenses):
    total_cost = 0
    for expense in expenses:
        total_cost += expense
    return total_cost

# 測試呼叫
print(daily_expenses(55, 120, 30)) # 買了三樣東西
print(daily_expenses())            # 今天沒花錢
```
