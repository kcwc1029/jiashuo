### Problem. 台北市 YouBike 2.0 騎乘費用計算
台北市 YouBike 2.0 目前的收費標準為：前 30 分鐘免費；超過 30 分鐘後，4 小時內每 30 分鐘 10 元（超過 30 分鐘的部分，未滿 30 分鐘以 30 分鐘計算）。請設計一個函數來計算預估費用。
* 函數名稱：calculate_youbike_fee
* 輸入變數：minutes (騎乘總分鐘數，整數)
* 輸出：回傳計算後的費用 (整數)
<!--  
```python=
import math

def calculate_youbike_fee(minutes):
    if minutes <= 30:
        return 0
    else:
        # 扣除免費的30分鐘後，計算需要收費的時間
        chargeable_minutes = minutes - 30
        # math.ceil 可以無條件進位，用來計算收費區間的數量
        intervals = math.ceil(chargeable_minutes / 30)
        fee = intervals * 10
        return fee

# 測試呼叫
result = calculate_youbike_fee(45)
print("騎乘45分鐘的費用為：", result, "元")
```

### Problem. 台灣銀行外幣兌換估算
出國旅遊前需要換匯，假設目前台灣銀行牌告匯率中，1 台幣大約可以兌換 4.71 日圓 (JPY) 或是 0.031 美元 (USD)。請設計一個函數，讓使用者輸入台幣預算後，同時得知可以換到多少日圓與美元。
* 函數名稱：exchange_currency
* 輸入變數：twd_amount (台幣金額，整數)
* 輸出：回傳兩個值，分別為日圓金額與美元金額 (浮點數，請使用 round 函數四捨五入到小數點後兩位)
<!--  
```python=
def exchange_currency(twd_amount):
    jpy_rate = 4.71
    usd_rate = 0.031
    
    jpy_amount = round(twd_amount * jpy_rate, 2)
    usd_amount = round(twd_amount * usd_rate, 2)
    
    return jpy_amount, usd_amount

# 測試呼叫並接收多個回傳值
jpy, usd = exchange_currency(10000)
```

### Problem. 台電住宅用電費估算
台灣電力公司的電價會區分「夏月」與「非夏月」。假設我們只計算單一級距（每月用電度數在 121 到 330 度之間），夏月每度電為 2.45 元，非夏月每度電為 2.16 元。請設計一個函數，根據用電度數同時回傳夏月與非夏月的預估電費。
* 函數名稱：estimate_electricity
* 輸入變數：kwh (用電度數，整數)
* 輸出：回傳兩個值，分別為夏月電費與非夏月電費 (整數，可直接使用 int() 轉換去除小數點)
<!--  
```python=
def estimate_electricity(kwh):
    summer_rate = 2.45
    non_summer_rate = 2.16
    
    summer_fee = int(kwh * summer_rate)
    non_summer_fee = int(kwh * non_summer_rate)
    
    return summer_fee, non_summer_fee

# 測試呼叫並接收多個回傳值
summer, non_summer = estimate_electricity(250)
print("用電250度，夏月電費約為：", summer, "元")
print("用電250度，非夏月電費約為：", non_summer, "元")
```
