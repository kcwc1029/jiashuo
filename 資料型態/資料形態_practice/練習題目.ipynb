### Problem. 資料的打包與拆包

變數名稱與輸入設定：
* secret_msg：數值設定為 "機密任務"
* encoded_msg：將 secret_msg 以 utf-8 編碼格式進行打包 (encode)
* decoded_msg：將 encoded_msg 解碼 (decode) 轉回字串

輸出要求：
* 第一行使用 len() 函式印出 encoded_msg 的位元組長度
* 第二行印出 decoded_msg 的內容
