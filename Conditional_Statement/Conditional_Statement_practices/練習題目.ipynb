### Problem. 珍珠奶茶購買判斷

一杯珍珠奶茶(boba_price)價格為 65 元。

請使用 `input()` 讓使用者輸入錢包餘額(wallet_money)，並轉換為整數。接著：

* 如果錢包餘額大於等於珍珠奶茶價格，輸出「餘額充足，可以購買」
* 否則輸出「餘額不足」



### Problem. 奇數與偶數判斷

請使用 `input()` 讓使用者輸入一個整數(number)，並轉換為整數。

接著：

* 如果 `number` 除以 2 的餘數等於 0，輸出「這是偶數」
* 否則輸出「這是奇數」

### Problem. 電影院票種判斷

請使用 `input()` 讓使用者輸入年齡(age)，並轉換為整數。

接著：

* 如果年齡小於 12，輸出「請購買兒童票」
* 如果年齡大於等於 12 且小於 65，輸出「請購買全票」
* 如果年齡大於等於 65，輸出「請購買敬老票」


### Problem. 百貨公司 VIP 滿額贈

百貨公司正在舉辦 VIP 滿額贈活動。

請使用 `input()` 讓使用者輸入消費金額(cost)，並轉換為整數。

VIP 會員(is_vip)設定為 `True`。

接著：

* 如果消費金額大於等於 1000，且為 VIP 會員，輸出「符合贈獎資格」
* 否則輸出「不符合贈獎資格」


### Problem. 遊樂園免費入場判斷

請使用 `input()` 讓使用者輸入身高(height)，並轉換為整數。

再使用 `input()` 讓使用者輸入年齡(age)，並轉換為整數。

接著：

* 如果身高小於 120，或年齡大於等於 70，輸出「免費入場」
* 否則輸出「請購票入場」


### Problem. 戶外體育課判斷

今天是否下雨(is_raining)設定為 `False`。

接著：

* 如果 `not is_raining` 為 `True`，輸出「今天上戶外體育課」
* 否則輸出「今天在室內上課」


### Problem. 字元類型判斷

請使用 `input()` 讓使用者輸入一個字元(char_val)。

接著：

* 如果字元為英文小寫 `a-z` 或英文大寫 `A-Z`，輸出「這是一個英文字母」
* 如果字元為數字 `0-9`，輸出「這是一個阿拉伯數字」
* 否則輸出「這是其他符號」


### Problem. 個人頭像檢查

使用者的個人頭像(profile_picture)設定為 `None`。

接著：

* 如果 `profile_picture` 為 `None`，輸出「使用者尚未上傳照片」
* 否則輸出「照片讀取成功」
